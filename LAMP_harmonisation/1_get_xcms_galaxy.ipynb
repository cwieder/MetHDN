```python
import pandas as pd
import requests
```

```python
def download_xcms_data(url_variableMetadata, url_dataMatrix, ID):
    vmd = pd.read_csv(url_variableMetadata, header=0, sep="\t")
    dm = pd.read_csv(url_dataMatrix, header=0, sep="\t")

    xcms_data = pd.merge(vmd, dm, on="name")
    xcms_data.to_csv(ID + "_xcms_data.tsv", index=False, sep="\t")
    return xcms_data
```

```python
download_xcms_data(
    url_variableMetadata='https://metabolights-labs.org/api/datasets/b73c104868c84afc/display?to_ext=tabular',
    url_dataMatrix='https://metabolights-labs.org/api/datasets/7ecb0e962e21d5e8/display?to_ext=tabular',
    ID='MTBLS3852'
)
```
